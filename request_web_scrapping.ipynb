### 4. POST

server address: https://httpbin.org/post

```python
def get_data_format(rcd_response):
    json_response = rcd_response.json()
    data_response = json_response['headers']['Content-Type'].split('/')[-1]
    print(f"Response data format: {data_response}")
```

```python
uri  = "https://httpbin.org/post"
```

```python
data = {
    "username":"bruce",
    "passowrd": "bruce123"
}
```

```python
response = requests.post(uri, data=data)
```

```python
response.status_code
```

```python
response.json()
```

```python
response.json()['headers']['Content-Type'].split('/')[-1]
```

```python
get_data_format(response)
```

### 9. Working with a public API
- API https://jsonplaceholder.typicode.com/
- endpoint: posts
- error handling

```python
url = "https://jsonplaceholder.typicode.com/posts"
```

```python
try:
    response = requests.get(url)
    response.raise_for_status()
except Exception as e:
    print(e)
else:
    status_code = response.status_code
    print(f"Status Code: {status_code}")
    if status_code == 200:
        print("\n Get request successful")
        posts = response.json()
        for i in range(3):
            print(f"\n Post {i + 1}")
            print(posts[i])
    else:
        print("Unsuccesful")
```

- Make a POST request to submit a post
- include following parameters
    - title
    - body
    - userid

```python
new_post = {
    "title":"Sample Post",
    "body":"This is a sample post",
    "userId":101
}
```

```python
try:
    response = requests.post(url,data=new_post)
    response.raise_for_status()
except Exception as e:
    print(e)
else:
    status_code = response.status_code
    print(f"Status Code: {status_code}")
    if status_code == 201:
        print("\n Post request successful")
        posts = response.json()
        print("\n Post :")
        print(posts)
    else:
        print("Unsuccesful")
```
